```python
import numpy as np
import matplotlib.pylab as plt
```

Step function

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
def step_function(x: np.array):
    y = x > 0
    return y.astype(np.int)
```

```python
x = np.array([-1.0, 1.0, 2.0])
step_function(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

Sigmoid function

$h(x) = \frac{1}{1 + \exp(-x)}$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

ReLU function

$h(x)=
\begin{cases}
x\;(x >0)\\
0\;(x \leq 0) 
\end{cases}$

```python
def relu(x):
    return np.maximum(x, 0)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.xlim(-5.5, 5.5)
plt.ylim(-0.5, 5.5)
plt.show()
```
